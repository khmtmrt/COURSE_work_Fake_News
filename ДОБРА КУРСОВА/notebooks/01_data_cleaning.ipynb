## 1. Завантаження даних

```python
# Завантажимо CSV файл з використанням абсолютного шляху
import os
from pathlib import Path

# Визначимо правильний шлях до кореневої папки проекту
project_root = Path(__file__).parent.parent if '__file__' in globals() else Path('..').resolve()
raw_data_path = project_root / 'data' / 'raw'

print(f"Робоча директорія: {os.getcwd()}")
print(f"Шлях до data/raw: {raw_data_path}")

csv_path = raw_data_path / 'news_classification_refined.csv'
print(f"Шлях до CSV файлу: {csv_path}")
print(f"Файл існує: {csv_path.exists()}")

df = pd.read_csv(csv_path)

print(f"\nЗавантажено {len(df)} записів")
print(f"Колонки: {df.columns.tolist()}")
display(df.head())
```

## 2. Інформація про датасет

```python
# Перевіримо наявність пропущених значень
print("Пропущені значення:")
print(df.isnull().sum())
```

## 3. Препроцесинг

```python
# Виконаємо базовий препроцесинг
print("Виконання препроцесингу...")
df_processed = preprocess_dataframe(df, text_column='text')

print(f"\nДо препроцесингу: {len(df)} записів")
print(f"Після препроцесингу: {len(df_processed)} записів")
```

```python
# Подивимось на приклади обробленого тексту
print("Приклади обробленого тексту:\n")
for i in range(3):
    print(f"\n--- Приклад {i+1} ---")
    print(f"Оригінал: {df_processed.iloc[i]['text'][:200]}...")
    print(f"Оброблений: {df_processed.iloc[i]['processed_text'][:200]}...")
```

## 4. Збереження обробленних даних

```python
# Збережемо оброблені дані
save_processed_data(df_processed, 'cleaned_data.csv')
print(f"\nОброблені дані збережено в {PROCESSED_DATA_PATH}")
```
